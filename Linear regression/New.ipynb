```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Independent variable (Study Hours) — must be 2D
X = np.array([[1], [2], [3], [4], [5]])

# Dependent variable (Marks)
y = np.array([45, 50, 55, 60, 65])
```

```python
# Create the model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)
```

```python
# Get the slope (coefficient) and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
# Predict marks using the same study hours
y_pred = model.predict(X)

# Print predictions
print("Predicted Marks:", y_pred)
```

```python
hours = 6
predicted_marks = model.predict([[hours]])
print(f"Predicted Marks for {hours} hours of study: {predicted_marks[0]}")
```

```python
# Scatter plot of actual data
plt.scatter(X, y, color='blue', label='Actual Marks')

# Line plot of predictions
plt.plot(X, y_pred, color='red', label='Regression Line')

# Labels and title
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression - Study Hours vs Marks")
plt.legend()
plt.grid(True)
plt.show()
```
