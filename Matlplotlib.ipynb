```python
import matplotlib.pyplot as plt
```

```python
a=[2,4,7,3,4,8,1]
labels=['A','B','C','D','E','F','G']
b=[1,4,6,4,2,4,1]
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(a,b,linestyle="--",marker=">")
plt.grid(True)
plt.show()
```

```python
plt.scatter(a,b,color="red")
plt.show()
```

```python
plt.hist(a,color='orange',bins=10)
```

```python
plt.bar(a,b)
```

```python
plt.pie(a,labels=labels,startangle=90)
plt.show()
```
