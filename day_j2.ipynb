```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.xlabel("x asis")
plt.ylabel("Y-aIXa")
plt.show()
```

```python
x=[1,2,3,4]
y=[5,2,1,8]
color1='r'
color2='b'
plt.plot(y,color1,marker='>')
plt.plot(x,color2,marker='o')

plt.show()
```

```python
age=[25,24,21,22,20]
reading_hours=[5,2,3,7,4]
plt.title("Age to Reading hour ratio")
plt.xlabel("Age")
plt.ylabel("reading_hours")
plt.bar(age,reading_hours,color="r")
plt.show()
```

```python
x=[2,3,4,5,1,3,7,8,9]
y=[1,2,3,4,5,6,7,8,9]
plt.title("Scatter Float")
plt.scatter(x,y)
plt.show()
```

```python
lbl=["Education","Food","Rent", "clothing","misc"]
x=[20,30,40,50,100]
plt.pie(x,labels=lbl) 
plt.show()
```
