```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("placement.csv")
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lac))')
plt.title('CGPA vs Placement Package')
plt.show()
```

```python
x=df[['cgpa']]
y=df[['package']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression   #importing linear regression method
LRinstance=LinearRegression()
```

```python
LRinstance.fit(x_train,y_train)
```

```python
print("Slope(m)",LRinstance.coef_)
print("intercept (b):",LRinstance.intercept_)
```

```python
y_pred=LRinstance.predict(x_test)  #Predict entire data
y_pred
```

```python
actual=y_test.values.ravel()
predicted=y_pred.ravel()
comparision_df=pd.DataFrame({
    'Actual_package':actual,
    'Predicted_package':predicted
})
```

```python
print(comparision_df.head())
```

```python
plt.figure(figsize=(8,6))
plt.scatter(actual,predicted,color='blue',label='predicted vs actual')
plt.plot([actual.min(),actual.max()],[actual.min(),actual.max()],'k--',lw=2,label='Perfect Prediction')
plt.xlabel('Actual Package(LPA)')
plt.ylabel('Predicted Package(LPA)')
plt.title('Actual vs Predicted:Scatter Plot')
plt.legend()
plt.grid(True)
plt.show()
```
